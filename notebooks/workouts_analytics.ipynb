```python
workout_type_map = {
    "HKWorkoutActivityTypeWalking": "walking",
    "HKWorkoutActivityTypeRunning": "running",
    "HKWorkoutActivityTypeCycling": "cycling",
    "HKWorkoutActivityTypeElliptical": "elliptical",
    "HKWorkoutActivityTypeFunctionalStrengthTraining": "strength_training",
    "HKWorkoutActivityTypeYoga": "yoga",
    "HKWorkoutActivityTypeSnowSports": "snow_sports",
    "HKWorkoutActivityTypeOther": "other",
}

df_workouts["workout_type"] = df_workouts["workoutActivityType"].map(workout_type_map)
```

```python
# Reorder columns to move workout_type before duration
cols = df_workouts.columns.tolist()

# Define desired order for the first few columns
new_order = (
    ["workout_type", "duration", "durationUnit"]
    + [c for c in cols if c not in ["workout_type", "duration", "durationUnit"]]
)

df_workouts = df_workouts[new_order]
```

```python
df_workouts["startDate"] = pd.to_datetime(df_workouts["startDate"])
df_workouts["endDate"] = pd.to_datetime(df_workouts["endDate"])
```

```python
df_zepp = df_workouts[df_workouts["sourceName"] == "Zepp"].copy()
df_strava = df_workouts[df_workouts["sourceName"] == "Strava"].copy()
```

```python
df_zepp["workout_date"] = df_zepp["startDate"].dt.date
df_strava["workout_date"] = df_strava["startDate"].dt.date
```

```python
merged = pd.merge(
    df_zepp,
    df_strava,
    on=["workout_type", "workout_date"],
    suffixes=("_zepp", "_strava")
)

# Duration difference in minutes
merged["duration_diff"] = abs(
    merged["duration_zepp"] - merged["duration_strava"]
)

# Consider duplicates if duration diff ≤ 2 minutes
duplicates = merged[merged["duration_diff"] <= 2]

len(duplicates)
```
