```python
sleep_value_map = {
    "HKCategoryValueSleepAnalysisInBed": "in_bed",
    "HKCategoryValueSleepAnalysisAwake": "awake",
    "HKCategoryValueSleepAnalysisAsleepCore": "asleep_core",
    "HKCategoryValueSleepAnalysisAsleepREM": "asleep_rem",
    "HKCategoryValueSleepAnalysisAsleepDeep": "asleep_deep",
    "HKCategoryValueSleepAnalysisAsleepUnspecified": "asleep_unspecified",
}

df_sleep["sleep_state"] = df_sleep["value"].map(sleep_value_map)
```

```python
df_sleep = df_sleep.drop(columns=["MetadataEntry", "type"])
```

```python
df_sleep["sourceName"] = df_sleep["sourceName"].replace({
    "iPhone (2)": "iPhone",
    "Zepp": "Amazfit",
    "Zepp Life": "Mi Smart Band",
    "Power Sync": "Fitbit"
})
```

```python
# Ensure datetime (should already be done, but safe)
df_sleep["startDate"] = pd.to_datetime(df_sleep["startDate"])
df_sleep["endDate"] = pd.to_datetime(df_sleep["endDate"])

# Compute duration in minutes
df_sleep["duration_min"] = (
    df_sleep["endDate"] - df_sleep["startDate"]
).dt.total_seconds() / 60
```

```python
df_sleep
```

```python
df_sleep["duration_min"].describe()
```

```python
(df_sleep["duration_min"] < 0).sum()
```

```python
#short sleep events
(df_sleep["duration_min"] < 1).sum()
```

```python
sleep_states_asleep = [
    "asleep_core",
    "asleep_rem",
    "asleep_deep",
    "asleep_unspecified"
]

df_sleep["is_asleep"] = df_sleep["sleep_state"].isin(sleep_states_asleep).astype(int)
```

```python
df_sleep["is_asleep"].value_counts()
```

```python
df_sleep["sleep_date"] = df_sleep["startDate"].dt.date
```

```python
df_sleep[["startDate", "sleep_date"]].head()
```

```python
# Total sleep per night (minutes)
nightly_sleep = (
    df_sleep[df_sleep["is_asleep"] == 1]
    .groupby("sleep_date")["duration_min"]
    .sum()
    .reset_index(name="total_sleep_min")
)
```

```python
nightly_sleep.describe()
```

```python
nightly_sleep.head()
```

```python
# Sleep stage breakdown per night
sleep_stages_nightly = (
    df_sleep[df_sleep["is_asleep"] == 1]
    .groupby(["sleep_date", "sleep_state"])["duration_min"]
    .sum()
    .unstack(fill_value=0)
    .reset_index()
)
```

```python
sleep_stages_nightly.head()
```

```python
sleep_stages_nightly.columns
```

```python
# Merge total sleep with stage breakdown
nightly_sleep_full = nightly_sleep.merge(
    sleep_stages_nightly,
    on="sleep_date",
    how="left"
)
```

```python
# Recompute total from stages
nightly_sleep_full["stage_sum_check"] = (
    nightly_sleep_full["asleep_core"]
    + nightly_sleep_full["asleep_rem"]
    + nightly_sleep_full["asleep_deep"]
    + nightly_sleep_full["asleep_unspecified"]
)
```

```python
# Difference between two totals
nightly_sleep_full["sleep_diff"] = (
    nightly_sleep_full["total_sleep_min"]
    - nightly_sleep_full["stage_sum_check"]
)
```

```python
nightly_sleep_full["sleep_diff"].describe()
```

```python
nightly_sleep_full
```

```python
nightly_sleep_full["sleep_extreme_flag"] = (
    (nightly_sleep_full["total_sleep_min"] < 60)
    | (nightly_sleep_full["total_sleep_min"] > 900)
).astype(int)

nightly_sleep_full["sleep_extreme_flag"].value_counts()
```

```python
# In-bed time per night
nightly_in_bed = (
    df_sleep[df_sleep["sleep_state"] == "in_bed"]
    .groupby("sleep_date")["duration_min"]
    .sum()
    .reset_index(name="in_bed_min")
)
```

```python
nightly_sleep_full = nightly_sleep_full.merge(
    nightly_in_bed,
    on="sleep_date",
    how="left"
)
```

```python
nightly_sleep_full["sleep_efficiency"] = (
    nightly_sleep_full["total_sleep_min"]
    / nightly_sleep_full["in_bed_min"]
)
```

```python
nightly_sleep_full.describe()
```

```python
nightly_sleep_full["sleep_efficiency"].describe()
```

```python
nightly_sleep_full["sleep_eff_invalid_flag"] = (
    (nightly_sleep_full["sleep_efficiency"] > 1)
    | (nightly_sleep_full["sleep_efficiency"] <= 0)
).astype(int)

nightly_sleep_full["sleep_eff_invalid_flag"].value_counts()
```

```python
nightly_sleep_full.loc[
    nightly_sleep_full["sleep_eff_invalid_flag"] == 1,
    "sleep_efficiency"
] = np.nan
```

```python
nightly_sleep_full["sleep_efficiency"].describe()
```

```python
nightly_sleep_full["deep_pct"] = (
    nightly_sleep_full["asleep_deep"] / nightly_sleep_full["total_sleep_min"]
)

nightly_sleep_full["rem_pct"] = (
    nightly_sleep_full["asleep_rem"] / nightly_sleep_full["total_sleep_min"]
)

nightly_sleep_full["core_pct"] = (
    nightly_sleep_full["asleep_core"] / nightly_sleep_full["total_sleep_min"]
)

nightly_sleep_full["unspecified_pct"] = (
    nightly_sleep_full["asleep_unspecified"] / nightly_sleep_full["total_sleep_min"]
)
```

```python
nightly_sleep_full
```

```python
nightly_sleep_full["poor_sleep_flag"] = (
    (nightly_sleep_full["total_sleep_min"] < 360) |
    (nightly_sleep_full["sleep_efficiency"] < 0.75)
).astype(int)
```

```python
nightly_sleep_full
```

```python
nightly_sleep_full["excellent_sleep_flag"] = (
    (nightly_sleep_full["total_sleep_min"] >= 420) &
    (nightly_sleep_full["sleep_efficiency"] >= 0.85)
).astype(int)
```

```python
nightly_sleep_full
```

```python
nightly_sleep_full["low_deep_sleep_flag"] = (
    nightly_sleep_full["deep_pct"] < 0.10
).astype(int)
```

```python
nightly_sleep_full["low_rem_sleep_flag"] = (
    nightly_sleep_full["rem_pct"] < 0.15
).astype(int)
```

```python
nightly_sleep_full["fragmented_sleep_flag"] = (
    (nightly_sleep_full["sleep_efficiency"] < 0.80) &
    (nightly_sleep_full["in_bed_min"] - nightly_sleep_full["total_sleep_min"] > 90)
).astype(int)
```

```python
nightly_sleep_full[
    ["deep_pct", "rem_pct", "core_pct", "unspecified_pct"]
].describe()
```

```python
nightly_sleep_full
```

```python
nightly_sleep_full["stage_available_flag"] = (
    nightly_sleep_full["asleep_unspecified"] < nightly_sleep_full["total_sleep_min"]
).astype(int)
```

```python
nightly_sleep_full
```

```python
sleep_export_cols = [
    "sleep_date",
    "total_sleep_min",
    "in_bed_min",
    "sleep_efficiency",

    "deep_pct",
    "rem_pct",
    "core_pct",
    "unspecified_pct",
    "stage_available_flag",

    "poor_sleep_flag",
    "excellent_sleep_flag",
    "low_deep_sleep_flag",
    "low_rem_sleep_flag",
    "fragmented_sleep_flag",
    "sleep_extreme_flag",
    "sleep_eff_invalid_flag"
]

daily_sleep_analytics = nightly_sleep_full[sleep_export_cols].copy()

daily_sleep_analytics.to_csv(
    "daily_sleep_analytics.csv",
    index=False
)
```
