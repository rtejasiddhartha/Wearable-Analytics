```python
def choose_canonical(group):
    interval_sources = group.dropna(subset=["intervals"])

    if not interval_sources.empty:
        return interval_sources.sort_values("intervals", ascending=False).iloc[0]
    else:
        return group.iloc[0]  # Fitbit-only day
```

```python
daily_steps_canonical = (
    daily_steps_by_source
    .groupby("date")[["sourceName", "total_steps", "intervals"]]
    .apply(choose_canonical)
    .rename(columns={
        "sourceName": "canonical_source",
        "total_steps": "canonical_steps"
    })
    .reset_index()
)
```
